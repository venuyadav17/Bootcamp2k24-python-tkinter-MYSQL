```python
# import pandas library
import pandas as pd

# load data
df = pd.read_csv('EDA_Loan.csv')
```

```python
# Indexing and Selection

#Access Column
print(df['Profession'])

# Access first row
print(df.loc[0])

# Access specific element
print(df.iloc[0, 1])
```

```python
# Data Manipulation

# Rename column

df = df.rename(columns={'Income (USD)': 'Salary'})

# Fill missing values
df['Salary'] = df['Salary'].fillna(0)

# Drop column
df = df.drop(columns=['Age'])
```

```python
#  Setting a Multi-Index

df.set_index(['Customer ID', 'Name'], inplace=True)
print(df)
```
