```python
#  Creating a Simple Line Plot

import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
# Customizing Plots

# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.plot(x, y1, label='Sine', color='blue', linestyle='--')
plt.plot(x, y2, label='Cosine', color='red', linestyle='-')

# Add title and labels
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Histogram

# Create data
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=30, edgecolor='black')

# Add title and labels
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Scatter Plot

# Create data
x = np.random.rand(100)
y = np.random.rand(100)

# Create the scatter plot
plt.scatter(x, y, color='green')

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
# Bar Plot

# Create data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create the bar plot
plt.bar(categories, values, color='purple')

# Add title and labels
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Subplots

# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, axs = plt.subplots(2)

# Plot on the first subplot
axs[0].plot(x, y1, label='Sine', color='blue')
axs[0].set_title('Sine Wave')
axs[0].legend()

# Plot on the second subplot
axs[1].plot(x, y2, label='Cosine', color='red')
axs[1].set_title('Cosine Wave')
axs[1].legend()

# Add overall title
fig.suptitle('Sine and Cosine Waves')

# Show the plots
plt.show()
```

```python
# example usage

import pandas as pd
import matplotlib.pyplot as plt

# Create sample data
data = {
    'Movie': ['Movie A', 'Movie B', 'Movie C', 'Movie D'],
    'Rating': [8.5, 7.2, 9.1, 6.8],
    'Year': [2020, 2019, 2018, 2021]
}

df = pd.DataFrame(data)

# Bar plot of movie ratings
plt.bar(df['Movie'], df['Rating'], color='skyblue')
plt.title('Movie Ratings')
plt.xlabel('Movie')
plt.ylabel('Rating')
plt.show()

# Scatter plot of ratings by year
plt.scatter(df['Year'], df['Rating'], color='red')
plt.title('Movie Ratings by Year')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.show()

# Line plot of ratings by year
plt.plot(df['Year'], df['Rating'], marker='o', linestyle='-', color='green')
plt.title('Movie Ratings Over Years')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.show()
```
